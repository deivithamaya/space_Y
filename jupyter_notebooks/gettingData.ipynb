```python
def getBoosterVersion(data)->list:
  """
    get the rocker's name for the id
  """
  url_rockets="https://api.spacexdata.com/v4/rockets/"

  #view if rockerList list exist
  #if ~("rockerList" in dir()): rockerList=[]
  rockerList=[]
  #request for all ids in data
  for x in data.rocket.values:
    responseRockets=requests.get(url_rockets+str(x)).json()
    rockerList.append(responseRockets["name"])

  return rockerList
```

```python
def getLaunchSite(data)->list:

  url_launches="https://api.spacexdata.com/v4/launchpads/"
  Longitude=[]
  Latitude=[]
  for x in data.launchpad.values:
    response=requests.get(url_launches+str(x)).json()
    Longitude.append(response["longitude"])
    Latitude.append(response["latitude"])
  return [Longitude,Latitude]
```

```python
def getPayloadData(data)->list:

  url_payload="https://api.spacexdata.com/v4/payloads/"
  PayloadMass=[]
  Orbit=[]

  for p in data.payloads.values:
    if len(p)==1:
      response=requests.get(url_payload+str(p[0])).json()
      PayloadMass.append(response["mass_kg"])
      Orbit.append(response["orbit"])
    else:
      PayForLarge=[]
      OrbitForLarge=[]

      for pp in p:
        response=requests.get(url_payload+str(pp)).json()
        PayForLarge.append(response["mass_kg"])
        OrbitForLarge.append(response["orbit"])

      PayloadMass.append(PayForLarge)
      Orbit.append(OrbitForLarge)

  return [PayloadMass, Orbit]
```
